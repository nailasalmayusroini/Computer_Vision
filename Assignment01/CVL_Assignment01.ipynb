```python
# Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Blurry Image

```python
# Problem: Blurry image
# Solution: Sharpening

image = cv2.imread("/content/street.jpg")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))


kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

sharpened = cv2.filter2D(image, -1, kernel)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Blurry Image")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Sharpened Image")
plt.imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

For blurry images, we use sharpening to enhance the image quality and to get more detail into what goes on in the image.

## Too bright / Too dark

```python
# Problem: Image too dark or too bright
# Solution: Brightness adjustment

image = cv2.imread("/content/too_dark.jpg")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))

brightness = 40
contrast = 1.2

brightness_fixed = cv2.addWeighted(image, contrast,
                                   np.zeros(image.shape, image.dtype),
                                   0, brightness)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Brightness Adjusted")
plt.imshow(cv2.cvtColor(brightness_fixed, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

In this case, I chose an image that is too dark and try to enhanc it. Here, to see the image better we adjust the brightness in the image so that if an image is too bright it can tone down the brightness and if it's too dark it can enhance the image's brightness

## Noisy Image

```python
# Problem: Noisy image
# Solution: Median filter

image = cv2.imread("/content/noisy.jpg")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))

denoised = cv2.medianBlur(image, 5)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Noisy Image")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Noise Removed")
plt.imshow(cv2.cvtColor(denoised, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

For noisy images, we use median filter to sort of smooth out the image. From the result of the image above, it can be seen that the noise that's was in the image is now gone and smoothed out. But we can also see that the image gets really blurry since we smoothed out the noise, so if we want to enhance it more we would need to sharpen the image.

## Low Contrast Image

```python
# Problem: Low contrast image
# Solution: Histogram Equalization

image = cv2.imread("/content/low_contrast.jpg")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

equalized = cv2.equalizeHist(gray)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Low Contrast")
plt.imshow(gray, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Histogram Equalized")
plt.imshow(equalized, cmap="gray")
plt.axis("off")

plt.show()
```

For low contrast images we use histogram equalization. From the image result, it can be seen that the trees are more visible now and we can see the whole image more clearly.

## Low Resolution Image

```python
# Problem: Low resolution
# Solution: Upsampling

image = cv2.imread("/content/distortion.jpg")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))

low_res = cv2.resize(image, (image.shape[1]//4, image.shape[0]//4))

upsampled = cv2.resize(low_res,
                       (image.shape[1], image.shape[0]),
                       interpolation=cv2.INTER_CUBIC)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Low Resolution")
plt.imshow(cv2.cvtColor(low_res, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Upsampled")
plt.imshow(cv2.cvtColor(upsampled, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

Here, we try to implement low resolution for the original image first, then use upsampling to make it better. I tries to use multiple images and I think that this one (that's also used in distortion) has the best result. It can be seen that the upsampled image is not exactly better than the original but I just wanted to highlight the upsample result from an image with low resolution.

## Too Sharp Image

```python
# Problem: Image too sharp
# Solution: Gaussian smoothing

image = cv2.imread("/content/too_sharp.webp")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))

smoothed = cv2.GaussianBlur(image, (9,9), 0)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Sharp Image")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Smoothed Image")
plt.imshow(cv2.cvtColor(smoothed, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

We can see from the original image that it's very sharp. So Gaussian Smoothing is used to smooth out the image and make it less sharp.

## Uneven Illumination

```python
# Problem: Uneven illumination
# Solution: CLAHE

image = cv2.imread("/content/uneven_illumination.jpeg")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_image = clahe.apply(gray)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Uneven Illumination")
plt.imshow(gray, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title("CLAHE Result")
plt.imshow(clahe_image, cmap="gray")
plt.axis("off")

plt.show()
```

The image that's used for this is a city light image that's captured from a distance. To see the difference in illumination more clearly, we turn it into a balck and white image first then use CLAHE to make the illumination mroe even. It can be seen that before only the city lights part are bright, the rest is mostly dark. But in the CLAHE result, it can be seen that the grass and the sky are mor evisible now because of the even out illumination

## Distorted Image

```python
# Problem: Image distortion
# Solution: Perspective correction

image = cv2.imread("/content/distortion.jpg")

# convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))

height, width = image.shape[:2]

pts1 = np.float32([[50,50],
                   [width-50,50],
                   [50,height-50],
                   [width-50,height-50]])

pts2 = np.float32([[0,0],
                   [width,0],
                   [0,height],
                   [width,height]])

matrix = cv2.getPerspectiveTransform(pts1, pts2)

corrected = cv2.warpPerspective(image, matrix, (width,height))

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Distorted Image")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Corrected Image")
plt.imshow(cv2.cvtColor(corrected, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

In image distortion, perspective correction is used to correct the distorted image. It can be seen from the original image that the windows were slightly curved and overall odd. But in the corrected image it can be seen that the windows are now straighten out and also zoomed in. This does not exactly make the image better in my opinion, but I just wanted to highlight how the perspective correction is able to correct an image's distortion in some parts.
